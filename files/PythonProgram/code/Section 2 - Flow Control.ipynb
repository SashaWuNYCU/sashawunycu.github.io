## Linear Congruential Generator
***
$X_{i+1} = a X_i + c \hbox{ (mod m)}$
### Two modulo-9 LCGs show how different parameters lead to different cycle lengths. 
### Each row shows the state evolving until it repeats. 

```python
def lcg(Xint,a,c,m,size):
    # Initialize the sequence array with zeros
    Xs = [Xint]
    # Generate the sequence
    for i in range(1,size):
        Xs.append((a * Xs[i - 1] + c) % m)
    return Xs
# Parameters for the LCG
seed = 1
a,c,m = 2,0,9
size = 12
# produce cycle of length 6
# Generate the sequence
random_seq = lcg(seed,a,c,m,size)
random_seq
```

```python
# Parameters for the LCG
seed = 3
a,c,m = 2,0,9
size = 4
# produce cycle of length 2
# Generate the sequence
random_seq = lcg(seed,a,c,m,size)
random_seq
```

```python
# Parameters for the LCG
seed = 0
a,c,m = 4,1,9
size = 18
# gives a cycle length of 9 with any seed in [0, 8]
# Generate the sequence
random_seq = lcg(seed,a,c,m,size)
random_seq
```

## The Estimation of $\pi$
***
Set $X=2U_1 -1$ and $Y = 2U_2 -1$   

Define 
$\displaystyle 
I = \begin{cases}
  1  & X^2+Y^2 \leq 1 \\
  0 & \text{otherwise}
\end{cases}
$  

$\displaystyle E(I) = P(X^2 + Y^2 \leq 1) = \frac{\pi}{4}$

```python
import numpy as np
n = 10000
runs = 10
for k in range(runs):
	U1 = np.random.uniform(0,1,n)
	U2 = np.random.uniform(0,1,n)
	X = 2 * U1 - 1
	Y = 2 * U2 - 1
	est = 4 * sum((X**2 + Y**2) <= 1) / n
	print(f'{k+1} round pi est = {est:.4}')
```

## Pricing Barrior Option : For Loop + nested if Loop
***
Consider a geometric brownian motion $\displaystyle \frac{dS_t}{S_t}=r dt+\sigma dW_t \hbox{ where } W_t \sim N(0,t)$  

Then the analytic solution is $\displaystyle S_T = S_0 e^{(r-\frac{\sigma^2}{2})t+\sigma \sqrt{T}Z} \hbox{ where } Z \sim N(0,1)$  

The final payoff of a call option is $\displaystyle e^{-rT}E[(S_T-K)^+]$

```python
from math import exp,sqrt
import numpy as np
#  warning
import warnings
warnings.filterwarnings('ignore',category = DeprecationWarning)
S0 = 100 # initial stock price
K = 110 # strike price
r = 0.03 # risk free rate
sigma = 0.25 # volatility
T = 0.5 # years to expiration
M = 100000 # iterations
saver = []
for i in range(M):
    St = S0 * exp((r - sigma**2 / 2) * T + sigma * sqrt(T) * np.random.randn(1))
    saver.append(exp(-r*T)*(St - K)*(St - K > 0))
print(f'option price = {sum(saver) / M:.3f}')
```

## Rejection Sampling : For Loop + nested If Loop
***
if the target density is $f(x)$, find a density $g(x)$ satisfy  

$f(x)\leq Mg(x) \hbox{ for some } M > 0$  

step 1 : generate $u^{(i)} \sim U(0,1)$ and $y^{(i)} \sim g(x)$  
step 2 : collect $(u^{(i)},y^{(i)})$ which satisfy $\displaystyle u^{(i)}\leq \frac{f(y^{(i)})}{Mg(y^{(i)})}$

## Example
***
$\displaystyle f(x) = \frac{1}{K\cosh(x^2-1)},~\hbox{where }K=\int_{\mathbf{R}}f(x)dx$  
$\displaystyle g(x) \sim N(0,1)$

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as scs
from scipy.integrate import quad
import warnings
warnings.filterwarnings('ignore',category = DeprecationWarning)
warnings.filterwarnings('ignore',category = RuntimeWarning)

#  normalized constant
normC = quad(lambda x:1 / np.cosh(x**2 - 1),-np.inf,np.inf)[0]

def target(x):
    return 1 / (normC * np.cosh(x**2 - 1))
def proposal(x):
    return scs.norm.pdf(x)

# rejection sampling
x0 = np.linspace(-4,4,801)
plt.figure(figsize = [8,6])
plt.plot(x0,target(x0),label = 'f(x)')
plt.plot(x0,proposal(x0),label = 'g(x)')
plt.plot(x0,1.8 * proposal(x0),label = 'Mg(x)')
plt.legend()
```

```python
runs = 5000
saver = []
for k in range(runs):
    uk = np.random.uniform()
    yk = np.random.normal()
    if uk <= target(yk) / (1.8 * proposal(yk)):
        saver.append(yk)
plt.hist(saver,density = True,bins = np.arange(-4,4,0.1))
plt.plot(x0,target(x0))
print(len(saver) / runs)
```

## Symmetric Random Work on $\mathbb{Z}^2$ : for loop + nested while loop with stopping rule
***
Let $\{X_i,i\in\mathbb{N}\}$ be a sequence of iid random variables satisfying  

$\forall i \in \mathbb{N}~~\forall (x,y)\in \{(0,1),(0,-1),(1,0),(-1,0)\}$  

$P(X_i=(x,y)) = \frac{1}{4}$  

Let $ S_n = \sum_{i=1}^{n}X_i$  

Then $S_n$ is called symmetric random walk on $\mathbb{Z}^2$

```python
import random
import matplotlib.pyplot as plt
np = 4 # No. of path
bdd = 25 
# filling the coordinates with random variables
plt.figure(figsize = [10,8])
for k in range(np):
    i = 1
    x = [0]
    y = [0]
    while i > 0:
        if (abs(x[i-1]) > bdd) | (abs(y[i-1]) > bdd):
            print(f'path {k} with steps = {i}')
            plt.plot(x,y,label = f'(path {k}, {i} steps)')
            break
        else:
            val = random.randint(1,4) # 4 direction
            if val == 1:
                tx,ty = x[i-1] + 1,y[i-1]
            elif val == 2:
                tx,ty = x[i-1] - 1,y[i-1]
            elif val == 3:
                tx,ty = x[i-1],y[i-1] + 1
            else:
                tx,ty = x[i-1],y[i-1] - 1
            x.append(tx)
            y.append(ty)
            i += 1
plt.xlim([-bdd,bdd])
plt.ylim([-bdd,bdd])
plt.legend(loc = 'best')
```

## EM Algorithm
***
### Suppose we generated a sequence of n random variables $Y_i \sim N(\theta,\sigma^2)$
### Imagine that we have only observed $Y_1,\cdots,Y_m$
### Intuitively, the estimation of $\theta$ should be the sample mean of m observations.

### Exact Solution $\hat{\theta} = \bar{Y}$

```python
# Exact Solution
n = 100
theta = 10
Y = np.random.normal(theta,1,n)
est1 = sum(Y) / len(Y)
print(f'exact solution = {est1:.4f}')
```

### Numerical Solution by minimize  $-\ln \mathcal{L}$

```python
# Numerical solution
from scipy.optimize import minimize
# define the negtive log likelihood
logL = lambda a: (100/2) * np.log(2 * np.pi) + (1/2) * sum([(yi - a)**2 for yi in Y])
minimize(logL,0.5)
```

## Incomplete Information
### Using EM algorithm

```python
Y_obs = Y[:50]
param_int = 0.5
runs = 50
tols = 1e-3
paramL = [param_int]
for r in range(runs):
    param_now = (sum(Y_obs) + (n - len(Y_obs)) * paramL[r]) / n
    paramL.append(param_now)
    print(f'iteration {r + 1}, theta = {paramL[r]:.4f}')
    if abs(param_now - paramL[r]) < tols:
        break
```

***
## Metropolis-Hastings Algorithm
***

## Problem Setup 
### Target distribution $\pi(x) = 3e^{\frac{-x^2}{2}} +e^{\frac{-(x-4)^2}{2}}$
### candidate generating density $q(x|x_t) \sim N(x_t,1)$

```python
# iteration steps
n = 5000
# Target dist
P = lambda x: 3 * np.exp(-x*x/2) + np.exp(-(x - 4)**2/2)
# Proposal dist
Q = lambda x, xt: np.exp(-(x - xt)**2/2)/(np.sqrt(2 * np.pi))
# Initial State
x0 = 0
# Current State
xt = x0
samples = []
# Perform M-H
for i in range(n):
    # Generate x' from q(x'|x)
	x_cand = np.random.normal(xt,1)
    # define alpha(y|x)
	alpha_prob = min(1,(P(x_cand) * Q(xt,x_cand)) / (P(xt) * Q(x_cand,xt)))
	ut = np.random.uniform(0,1)
	if ut < alpha_prob:
		xt = x_cand
	samples.append(xt)
samples = np.array(samples)
plt.hist(samples,bins = 50,density = True)
# draw the density curve
xg = np.linspace(-10, 10, 1000)
yg = P(xg)
normC = quad(P,-np.inf,np.inf)[0]
plt.plot(xg,yg / normC,label = 'Target density P(x)')
plt.title('Metropolis Hastings')
plt.legend(loc = 'best',shadow = True)
```
